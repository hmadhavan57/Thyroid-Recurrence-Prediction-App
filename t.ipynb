```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

encoder = OneHotEncoder(sparse_output=False, drop='first')
le = LabelEncoder()
```

```python
X=df.drop(columns='Recurred').reset_index(drop=True)
y=df['Recurred']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```

```python
import streamlit as st
```

```python
# Extract the feature names the model was trained on
trained_feature_names = model.feature_names_in_

def user_input_features():
    age = st.number_input('Age', min_value=0, max_value=100, value=30)
    gender = st.selectbox('Gender', df['Gender'].unique())
    smoking_history = st.selectbox('Smoking History', df['Smoking History'].unique())
    radiotheraphy_history = st.selectbox('Radiotheraphy History', df['Radiotheraphy History'].unique())
    
    # Thyroid Function Options
    thyroid_function = st.selectbox(
        'Thyroid Function',
        ['Clinical Hypothyroidism', 'Euthyroid', 'Subclinical Hyperthyroidism', 'Subclinical Hypothyroidism']
    )
    
    physical_examination = st.selectbox(
        'Physical Examination',
        ['Multinodular goiter', 'Normal', 'Single nodular goiter-left', 'Single nodular goiter-right']
    )
    
    adenopathy = st.selectbox(
        'Adenopathy',
        ['Extensive', 'Left', 'No', 'Posterior', 'Right']
    )
    
    pathology = st.selectbox(
        'Pathology',
        ['Hurthel cell', 'Micropapillary', 'Papillary']
    )
    
    focality = st.selectbox('Focality', df['Focality'].unique())
    risk = st.selectbox('Risk', df['Risk'].unique())
    tumor_classification = st.selectbox('Tumor Classification', df['Tumor Classification'].unique())
    nodal_classification = st.selectbox('Nodal classification', df['Nodal classification'].unique())
    metastasis_classification = st.selectbox('Metastasis classification', df['Metastasis classification'].unique())
    stage = st.selectbox('Stage', df['Stage'].unique())
    response = st.selectbox(
        'Response',
        ['Excellent', 'Indeterminate', 'Structural Incomplete']
    )

    # Create a dictionary with the selected options
    data = {
        'Age': age,
        'Gender': gender,
        'Smoking History': smoking_history,
        'Radiotheraphy History': radiotheraphy_history,
        f'Thyroid Function_{thyroid_function}': 1,
        f'Physical Examination_{physical_examination}': 1,
        f'Adenopathy_{adenopathy}': 1,
        f'Pathology_{pathology}': 1,
        'Focality': focality,
        'Risk': risk,
        'Tumor Classification': tumor_classification,
        'Nodal classification': nodal_classification,
        'Metastasis classification': metastasis_classification,
        'Stage': stage,
        f'Response_{response}': 1
    }
    
    # Add columns not selected by the user with value 0
    for col in trained_feature_names:
        if col not in data:
            data[col] = 0

    # Create DataFrame with columns in the same order as the training data
    input_df = pd.DataFrame(data, index=[0])[list(trained_feature_names)]
    
    return input_df

input_df = user_input_features()

# Predict outcome
prediction = model.predict(input_df)
prediction_proba = model.predict_proba(input_df)

st.subheader('Prediction')
result = 'Recurred' if prediction[0] else 'Did not Recur'
st.write(result)

st.subheader('Prediction Probability')
st.write(prediction_proba)

# Display model accuracy
st.subheader('Model Accuracy')
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
st.write(f'Accuracy: {accuracy:.2f}')
```
